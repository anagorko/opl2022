# Lemat Farkasa

Celem notatek jest pokazanie dowodu silnego twierdzenia o dualności za pomocą lematu Farkasa.

**Lemat Farkasa**.

Niech $A \in M_{m \times n}(\mathbb{R})$ będzie macierzą o współczynnikach rzeczywistych, $m$ wierszach i $n$ kolumnach. Niech $b \in \mathbb{R}^m$ będzie dowolnym wektorem. Wtedy zachodzi dokładnie jeden z warunków

**F1)** Istnieje wektor $x \in \mathbb{R}^n$ taki, że $Ax = b$ i $x \geq 0$.

**F2)** Istnieje wektor $y \in \mathbb{R}^m$ taki, że $y^T A \geq 0^T$ i $y^T b < 0$.

**Uwaga**. F1) i F2) nie mogą zachodzić równocześnie. Rzeczywiście, warunek F2) mówi, że istnieje kombinacja liniowa wierszy macierzy $A|b$ (macierzy $A$ uzupełnionej o kolumnę $b$) taka, że

$$
\left[
\begin{array}{ccccc|c}
  \vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
  \geq 0 & \geq 0 & \geq 0 & \ldots & \geq 0  & < 0 \\
  \vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
\end{array}
\right]
$$

To znaczy, że układ równań $Ax = b$ nie może być spełniony dla żadnego $x \geq 0$.

**Interpretacja logiczna**.

Pokażę na przykładzie. Dany jest układ nierówności

$$
\left\{
\begin{array}{rcl}
4x_1 + x_2 & \leq & 4 \\
-x_1 + x_2 & \leq & 1 \\
-2x_1 - x_2 & \leq & -3 \\
\end{array}
\right.
$$

Jeżeli dobierzemy nieujemne współczynniki $\alpha, \beta, \gamma \geq 0$, to możemy przemnożyć przez nie nasze nierówności (nie zmieniając znaku) i dodać je stronami.

Weźmy $\alpha = 3$, $\beta = 2$, $\gamma = 5$. Mamy

$$
\left\{
\begin{array}{rcl}
12x_1 + 3x_2 & \leq & 12 \\
-2x_1 + 2x_2 & \leq & 2 \\
-10x_1 - 5x_2 & \leq & -15 \\
\end{array}
\right.
$$

Po dodaniu stronami otrzymujemy

$$
0 + 0 \leq -1
$$

co pokazuje, że dany układ nierówności był sprzeczny. Lemat Farkasa mówi, że jeżeli układ jest sprzeczny, to zawsze możemy tak dobrać współczynniki $\alpha, \beta, \gamma$, by w powyższy sposób tę sprzeczność wykazać.

**Stwierdzenie**. (Lemat Farkasa geometrycznie)

Niech $a_1, a_2, \ldots, a_n, b$ będą wektorami w $\mathbb{R}^n$. Zachodzi dokładnie jeden z poniższych warunków:

**F1')** Punkt $b$ leży w stożku wypukłym $C$ generowanym przez $a_1, a_2, \ldots, a_n$.

(Stożek generowany przez $a_1, a_2, \ldots, a_n$ to zbiór

$$
\{ t_1 a_1 + t_2 a_2 + \cdots + t_n a_n \colon t_1, t_2, \ldots, t_n \geq 0 \}
$$

czyli otoczka wypukła sumy półprostych $\overrightarrow{0a_i}$ po $i = 1, 2, \ldots, n$)

**F2')** Istnieje hiperpłaszczyzna przechodząca przez $0$ postaci $h = \{ x \in \mathbb{R}^m \colon y^T x = 0 \}$ dla pewnego $y \in \mathbb{R}^m$ taka, że wektory
$a_1, a_2, \ldots, a_n$ (i cały stożek $C$) leżą po jednej jej stronie a wektor $b$ po drugiej, tzn.

$$
y^T b < 0 \text{ i } y^T a_i \geq 0 \text{ dla } i = 1, 2, \ldots n.
$$

 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;


**Lemat** (Farkasa w trzech wariantach)

Niech $A \in M_{m \times n}(\mathbb{R})$ i $b \in \mathbb{R}^m$.

**i)** $Ax = b$ ma nieujemne rozwiązanie $x \geq 0$ jedynie jeśli dla każdego $y \in \mathbb{R}^m$ takiego, że $y^T A \geq 0^T$ mamy $y^T b \geq 0$.

**ii)** $Ax \leq b$ ma nieujemne rozwiązanie $x \geq 0$ jedynie jeśli dla każdego nieujemnego $y \in \mathbb{R}^m$, $y \geq 0$, takiego, że $y^T A \geq 0^T$ mamy $y^T b \geq 0$.

**iii)** $Ax \leq b$ ma rozwiązanie $x$ jedynie jeśli dla każdego nieujemnego $y \in \mathbb{R}^m$, $y \geq 0$ takiego, że $y^T A = 0^T$ mamy $y^T b \geq 0$.


Pokażę implikacje 

i) $\Rightarrow$ ii) $\Leftarrow$ iii) 

bo w dowodzie silnego twierdzenia o dualności wykorzystam ii). Pozostałe implikacje

i) $\Leftarrow$ ii) $\Rightarrow$ iii)

pozostawiam jako ćwiczenie.

**Dowód** i) $\Rightarrow$ ii).

Lewa strona ii): $Ax \leq b$ ma rozwiązanie $x \geq 0$.

Niech $\widetilde A = (A | I_m)$, gdzie $I_m$ to macierz jednostkowa $m \times m$.

Zauważmy, że $Ax \leq b$ ma rozwiązanie $x \geq 0$ jedynie jeśli istnieje $\tilde x \geq 0$ takie, że $\widetilde A \tilde x = b, \tilde x \geq 0$.
Z i) jest to równoważne temu, że dla każdego $y$ jeżeli $y^T \widetilde A \geq 0$, to $y^T b \geq 0$.
Ale $y^T \widetilde A \geq 0$ to to samo, co $y^T A \geq 0$ i $y \geq 0$. A to jest prawa strona ii).




**Dowód** iii) $\Rightarrow$ ii).

Lewa strona ii): $Ax \leq b$ ma rozwiązanie $x \geq 0$.

Niech $\widetilde A = \begin{pmatrix}A\\ \hline -I_n\end{pmatrix}$, $\tilde b = \begin{pmatrix}b \\ \hline 0\end{pmatrix}$.

Zauważmy, że $Ax \leq b$ ma rozwiązanie $x \geq 0$ jedynie jeśli $\widetilde A x \leq \tilde b$ ma dowolne rozwiązanie.
Z iii) jest to równoważne temu, że dla każdego $\tilde y \geq 0$ takiego, że $\tilde y^T \widetilde A = 0$ mamy $\tilde y^T \tilde b \geq 0$.

Niech $\tilde y = \begin{pmatrix}y \\ \hline y'\end{pmatrix}$. Mamy $\tilde y \geq 0$, $\tilde y^T \widetilde A = 0^T$ jedynie jeśli
$y \geq 0$, $y'^T = y^T A \geq 0^T$ i $\tilde y^T \tilde b = y^T b$. Stąd $Ax \leq b$ ma nieujemne rozwiązanie $x \geq 0$ jedynie jeśli 
dla każdego $y \geq 0$ takiego, że $y^T A \geq 0$ mamy $y^T b \geq 0$.


----------

Tezy lematu Farkasa można podsumować w następującej tabeli

$$
\begin{array}{r|c|c}
 & \text{Układ } Ax \leq b & \text{Układ } Ax = b \\
 \hline
 \text{ma rozw. } x \geq 0 \Leftrightarrow & \forall y: y \geq 0, y^T A \geq 0 \Rightarrow y^Tb \geq 0 & \forall y: y^T A \geq 0^T \Rightarrow y^T b \geq 0 \\
 \hline
 \text{ma rozw. } x \in \mathbb{R}^n \Leftrightarrow & \forall y: y \geq 0, y^T A = 0 \Rightarrow y^T b \geq 0 & \forall y: y^T A = 0^T \Rightarrow y^T b = 0 \\
\end{array}
$$

Proszę zwrócić uwagę na "brakującą" kombinację z prawego dolnego rogu (nie mamy warunku iv)): jest to po prostu eliminacja Gaussa.

**Interpretacja logiczna**.

Jeżeli układ nierówności $Ax \leq b$ jest sprzeczny, to wykonując elementarne operacje na wierszach (zachowujące znaki nierówności) można doprowadzić układ do nierówności $0 \leq -1$.

Dowód z iii): $Ax \leq b$ jest sprzeczny $\Leftrightarrow$ istnieje $y \geq 0$ takie, że $y^T A = 0^T$ i $y^T < b$. Ale $y^T$ to współczynniki do kombinacji liniowej wierszy układu, która daje nierówność $0 \leq y^T b < 0$. Dzieląc te współczynniki przez $|y^T b|$ uzyskujemy nierównosć $0 \leq -1$.


**Dowód silnego twierdzenia o dualności** z ii).

(Silne twierdzenie o dualności:

Dla programów liniowych

(P) $c^T x \to \max \colon Ax \leq b, x \geq 0$

i

(D) $b^T y \to \min \colon A^Ty \geq c, y \geq 0$

zachodzi dokładnie jedna z możliwości

1. Zarówno (P) jak i (D) są sprzeczne.
2. (P) jest nieograniczony i (D) jest sprzeczny.
3. (P) jest sprzeczny i (D) jest nieograniczony.
4. (P) i (D) mają rozwiązania dopuszczalne (są niesprzeczne). Wtedy oba mają rozwiązania optymalne i jeżeli $x^\ast$ jest rozwiązaniem optymalnym (P) a $y^\ast$ jest rozwiązaniem optymalnym (D),
to 

$$c^T x^\ast = b^T y^\ast$$

(Czyli maksimum (P) jest równe minimum (D).)

Załóżmy, że (P) ma rozwiązanie optymalne $x^\ast$. Niech $\gamma = c^T x^\ast$ - optymalna wartość funkcji celu dla (P). Wiemy, że układ

$$
Ax \leq b, c^T x \geq \gamma \quad\quad (*)
$$

ma nieujemne rozwiązanie, ale dla dowolnego $\varepsilon > 0$ układ

$$
Ax \leq b, c^T x \geq \gamma + \varepsilon \quad\quad (**)
$$

nie ma nieujemnych rozwiązań.

Definujemy macierz $\widetilde A \in M_{(m+1) \times n}(\mathbb{R})$ i wektor $\tilde b_\varepsilon \in \mathbb{R}^{m+1}$ przez

$$\widetilde A = \begin{pmatrix}A \\ \hline -c^T\end{pmatrix}, \tilde b_\varepsilon = \begin{pmatrix}b \\ \hline -\gamma - \varepsilon\end{pmatrix}.$$

Wtedy $(*)$ jest równoważne $\widetilde A x \leq \tilde b_0$ i $(**)$ jest równoważne $\widetilde A x \leq \tilde b_\varepsilon$.

Stosujemy wariant ii) lematu Farkasa. Dla $\varepsilon > 0$ wnioskujemy, że istnieje nieujemny wektor $\tilde y = (u, z) \in \mathbb{R}^m \times \mathbb{R}$ taki, że
$\tilde y^T \widetilde A \geq 0^T$ ale $\tilde y^T \tilde b_\varepsilon < 0$.

Stąd $A^T u \geq zc$, $b^T u < z(\gamma + \varepsilon)$.


Dla $\varepsilon = 0$ zastosowanie ii) pokazuje, że ten sam wektor $\tilde y$ spełnia $\tilde y^T \tilde b_0 \geq 0$ (bo układ ma nieujemne rozwiązanie).

Jest to równoważne temu, że $b^T u \geq z \gamma$. Stąd $z > 0$ bo $z = 0$ przeczy nierówności $b^T u < z(\gamma + \varepsilon)$ a $\tilde y \geq 0$.

Weźmy $v = \frac 1z u \geq 0$. Wtedy mamy $A^T v \geq c$ i $b^T v < \gamma + \varepsilon$, czyli $v$ jest rozwiązaniem dopuszczalnym (D) z wartością funkcji celu mniejszą niż $\gamma + \varepsilon$.
Ze słabego twierdzenia o dualności każde rozwiązanie dopuszczalne (D) daje wartość funkcji celu $\geq \gamma$. Stąd (D) jest ograniczony i ma jakieś rozwiązanie optymalne $y^\ast$. Oczywiście

$$
\gamma \leq b^T y^\ast \leq \gamma + \varepsilon
$$

dla każdego $\varepsilon > 0$, co kończy dowód.
